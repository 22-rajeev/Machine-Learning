```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generating senthetic data points:
X = 2*np.random.rand(100 ,1)
```

```python
# np.random.rand(100, 1)` - This generates a NumPy array containing 100 random numbers between 0 and 1, arranged in a column vector (100 rows, 1 column).

# `2*np.random.rand(100, 1)` - This multiplies each random value by 2, scaling the range from [0,1] to [0,2].

# The result is assigned to variable `X`, which will be a 100×1 array of random values between 0 and 2.
```

```python
y = 4 +3*X +1.5*X**2 + np.random.randn(100,1)
```

```python
# train test split 
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size=0.2 , random_state=1)
```

```python
X_train.shape , X_test.shape , y_train.shape , y_test.shape
```

```python
from sklearn.preprocessing import PolynomialFeatures 
```

```python
poly_features = PolynomialFeatures(degree=2 , include_bias=False)
X_poly_train = poly_features.fit_transform(X_train)
```

```python
X_poly_train
```

```python
from sklearn.linear_model import LinearRegression
poly_reg = LinearRegression()
```

```python
poly_reg
```

```python
poly_reg.fit(X_poly_train , y_train)
```

```python
poly_reg.coef_
```

```python
poly_reg.intercept_
```

```python
y_poly_predict = poly_reg.predict(X_poly_train)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_train , y_poly_predict)
```

```python
plt.scatter(X_train,y_train ,label ="Train Data")
```

```python
X_range = np.linspace(0, 2, 100).reshape(-1, 1)	
     # np.linspace(0, 2, 100) creates 100 evenly spaced numbers between 0 and 2.
	# reshape(-1, 1) converts it into a column vector with shape (100, 1), which is required for sklearn models.

plt.scatter(X_train,y_train ,label ="Train Data")

X_range_poly = poly_features.transform(X_range)

plt.plot(X_range, poly_reg.predict(X_range_poly), color='r', label='Polynomial Regression')
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```
