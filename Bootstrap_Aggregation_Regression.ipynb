```python
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
```

```python
X , y = make_regression(n_samples=1000 , n_features=10 , noise=0.1 ,random_state=True)
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y , test_size=0.35 , random_state=True)
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LinearRegression 
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR 
```

```python
lr = LinearRegression()
dtr = DecisionTreeRegressor()
svr = SVR()
```

```python
from sklearn.ensemble import VotingRegressor
```

```python
ensemble_regressor = VotingRegressor(estimators=[('lr',lr) ,('dtr',dtr) , ('svr' , svr)])
```

```python
ensemble_regressor
```

```python
ensemble_regressor.fit(X_train ,y_train)
```

```python
y_pred = ensemble_regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test ,y_pred)
```
