```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_diabetes 
```

```python
data = load_diabetes()
```

```python
# Problem statement : To predict disease progression one year after baseline provided 10 freatures
```

```python
print(data.DESCR)
```

```python
data.data
```

```python
data.target
```

```python
data.feature_names
```

```python
pd.DataFrame(data.data , columns = data.feature_names) # Creating a pandas dataframe from the given scikit learn data
```

```python
# Sex has a value in decimals which means it has been scaled out and it is done in order to hide personal data of people.
```

```python
Model_data = pd.DataFrame(data.data , columns = data.feature_names)
```

```python
Model_data.head()
```

```python
Model_data['target']=data.target
```

```python
Model_data.head()
```

```python
# After this step we usually do EDA , Datacleaning , NULL value Treatment since data is from scikit learn it is already done for us.
```

```python
X = Model_data[['bmi']] # independent variable
Y = Model_data['target'] # dependent variable
print(X,Y)
```

```python
# Train Test Split 
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , Y_train , Y_test = train_test_split(X , Y , test_size=0.2 ,random_state=1) 
# random_state :  Controls the shuffling applied to the data before applying the split. reproduces same result everytime
```

```python
print(X_train.shape , X_test.shape , Y_train.shape , Y_test.shape)
```

```python
# Sacaling is optional
```

```python
# Model Traning
```

```python
from sklearn.linear_model import LinearRegression 
```

```python
model = LinearRegression()
```

```python
 model #empty instance of Linear Regression
```

```python
# Traning of model : at backend it will use gradient descent to give optimal coeffients
model.fit(X_train , Y_train)
```

```python
# color changed blue means fitted
```

```python
model.coef_ # m it is an attribute of the model
```

```python
model.intercept_
```

```python
# Now we have to test the data on X_test and this will give us Y_predicted and we will then compare it to Y_test 
# inorder to check the performance of our model
```

```python
Y_predict = model.predict(X_test)
```

```python
Y_predict
```

```python
#visualize
```

```python
plt.scatter(X_test , Y_test , color = 'black' , label ='Actual Data')
plt.plot(X_test , Y_predict , color = 'blue' , linewidth = 3 , label ='Predicted Linear Regression Line')
plt.xlabel('BMI')
plt.ylabel('One Year Progression Target')
plt.title('Linear Regression on Diabetes Data')
plt.legend()
plt.show()
```

```python
#model evaluation
```

```python
from sklearn.metrics import mean_squared_error 
```

```python
mean_squared_error(Y_test , Y_predict)
```
