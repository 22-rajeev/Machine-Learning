```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
from sklearn.datasets import fetch_california_housing
```

```python
data = fetch_california_housing()
```

```python
print(data.DESCR)
```

```python
# problem statement: To predict the median house value for California District
```

```python
data.keys()
```

```python
data.target_names
```

```python
data.feature_names
```

```python
# Final data
df = pd.DataFrame(data.data , columns = data.feature_names)
```

```python
df
```

```python
df['target'] = data.target
```

```python
df
```

```python
df.describe() # 5 point summary
```

```python
df.corr() # correlation between features
```

```python
sns.heatmap(df.corr() ,annot=True)
```

```python
# seperate  X and y from df
X = df.iloc[:,:-1] # independent variable
y = df.iloc[:,-1] # target_variable
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train ,X_test, y_train , y_test =  train_test_split(X , y , test_size=0.25 , random_state=1)
```

```python
X_train.shape , X_test.shape , y_train.shape ,y_test.shape
```

```python
# sacling is optional 
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model
```

```python
model.fit(X_train , y_train)
```

```python
len(model.coef_)
```

```python
model.intercept_
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error , r2_score
```

```python
mean_squared_error(y_test , y_pred)
```

```python
r2_score(y_test , y_pred)
```

```python
error = y_test-y_pred
```

```python
sns.distplot(error)
```

```python
plt.scatter(y_test , y_pred)
```

```python
plt.scatter(y_pred , error) # homoscedasticity >> variance between y_pred and error should be constant
```

```python
# This model looks good...
```

```python
# Pickling is the process of serializing a trained model in Python so you can save it to a file and load it later without retraining.
```

```python
# Python pickle module is used for serializing or de-serialzing a python object or structure. Any object in python can be pickled.
# SERIALIZATION: It is the process of converting a data structure or object into a format that can easily be stored,reterieved and persisted.
# During serialization the objects state is converted into a stream of bytes or other format that can be written to disk.
```

```python
model
```

```python
import pickle
```

```python
pickle.dump(model , open("model.pkl" ,"wb")) # Serialization
```

```python
pickle.load(open("model.pkl" , "rb")) # Deserialization
```
