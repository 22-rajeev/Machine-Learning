```python
# Classifying IRISH flower as per petal length sepal length and width.
```

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
data
```

```python
data.feature_names
```

```python
df = pd.DataFrame(data= data.data , columns=data.feature_names)
```

```python
df
```

```python
df['target'] = data.target
```

```python
df
```

```python
df = df[df.target !=2]
```

```python
df
```

```python
x = df.iloc[:,:-1]
```

```python
y = df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(x , y , test_size= 0.20 , random_state=1)
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression()
classifier
```

```python
classifier.fit(X_train , y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
#logistic Regression gives us the probablility we can decide the cut-off for class 1 and class 2 in Logistic regression.
```

```python
classifier.predict_proba(X_test)
```

```python
# Above are probability of class 0 and class 1 in the 2d array
```

```python
# Evaluation matrix
from sklearn.metrics import confusion_matrix , accuracy_score , classification_report
```

```python
confusion_matrix(y_test , y_pred)
```

```python
accuracy_score(y_test , y_pred)
```

```python
print(classification_report(y_test , y_pred))
```

```python
# macro avg >> simple average precision and recall for class 0 and class 1
# support will be total number of data points
# weighted avg is used in case of class imbalance gives importance to performance of dominant class.
```
