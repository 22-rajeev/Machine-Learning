```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_classification 
# The `make_classification` function is used to generate synthetic datasets 
# for classification problems. It creates random samples with specified 
# features that can be used for testing and demonstrating classification algorithms.

X , y = make_classification(n_samples=1000 , n_features=10 , n_redundant=5 , n_informative= 5 , n_classes= 2, random_state=1)
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X ,y , test_size=0.3 , random_state=1)
```

```python
X_train.shape #70%
```

```python
X_test.shape #30%
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model
```

```python
model.fit(X_train , y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix , accuracy_score , classification_report
```

```python
print(classification_report(y_test , y_pred))
```

```python
confusion_matrix(y_test , y_pred)
# Calculate the confusion matrix to evaluate classification performance
# This compares the actual values (y_test) with the predicted values (y_pred)
# The matrix shows true positives, false positives, true negatives, and false negatives
```

```python
accuracy_score(y_test , y_pred)
```

```python
# Import necessary metrics from scikit-learn for ROC curve analysis
from sklearn.metrics import roc_curve, auc

# Generate probability predictions for the positive class (class 1)
# model.predict_proba returns probabilities for all classes, so we select only the second column [1]
y_pred_proba = model.predict_proba(X_test)[:,1]
# Display the predicted probabilities
y_pred_proba
```

```python
fpr , tpr , thresholds = roc_curve(y_test , y_pred_proba)

roc_auc = auc(fpr , tpr)
plt.figure(figsize=(8, 6))
# Plot ROC curve
plt.plot(fpr, tpr, color='darkorange', linewidth=2, label='ROC curve (area = %0.2f)' % roc_auc)

# Plot diagonal line (random chance)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
# Axis limits
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])

# Labels and title
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')

# Legend
plt.legend(loc="lower right")

# Show plot
plt.show()
```

```python
# by default the precision class is decided with cuttoff 0.5
# if we want to decide own cutoff we need to use precision, recall , accuracy tradeoff 
# sensistivity specificity accuracy tradeoff
```

```python
from sklearn.metrics import precision_score , accuracy_score , recall_score

thresholds = np.linspace(0,1,100)
precisions =[]
recalls=[]
accuracies=[]

for threshold in thresholds:
    y_pred_threshold = (y_pred_proba >= threshold).astype(int)
    precision = precision_score(y_test , y_pred_threshold)
    recall = recall_score(y_test , y_pred_threshold)
    accuracy = accuracy_score(y_test , y_pred_threshold)
    precisions.append(precision)
    recalls.append(recall)
    accuracies.append(accuracy)

plt.figure(figsize=(10,6))
plt.plot(thresholds , precisions , label ='Precision')
plt.plot(thresholds , accuracies , label ='Accuracy')
plt.plot(thresholds , recalls , label ='Recall')
plt.xlabel("Threshold Probabilities")
plt.ylabel("Score")
plt.title("Precision , Recall , Accuracy vs Threshold Probability")
plt.legend()
plt.grid(True)
plt.show()
```

```python
new_pred_levels = np.where(y_pred_proba >0.4 ,1,0)
```

```python
new_pred_levels
```

```python
accuracy_score(y_test , new_pred_levels)
```

```python
# Hence accuracy score of the overall model increased.
print(classification_report(y_test,new_pred_levels))
```
